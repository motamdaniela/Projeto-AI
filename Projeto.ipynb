```python
#imports

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score

from scipy import stats
```

```python
df = pd.read_csv('covid19.csv')
```

```python
df.isnull().sum()
```

### a lot of missing values

```python
# replacing the missing values with a 0
df = df.fillna(0)
```

```python
df.isnull().sum()
```

### all the missing values have been replaced

```python
#check out our table
pd.set_option('display.max_columns', None)
df.head()
```

### there are some values that should be turned to bolean(sex, patient type and date died)

```python
#replace the female value with a 1 and male with 0 in a new column
for index, value in df['SEX'].iteritems():
    if value == "male":
        df.at[index, 'GENDER'] = 0
    else:
        df.at[index, 'GENDER'] = 1

#do the same for patient type
for index, value in df['PATIENT_TYPE'].iteritems():
    if value == "returned home":
        df.at[index, 'HOSPITALIZATION'] = 0
    else:
        df.at[index, 'HOSPITALIZATION'] = 1
        
#for date died replace the '9999-99-99' default value for patients who didnt die of covid
for index, value in df['DATE_DIED'].iteritems():
    if value == "9999-99-99":
        df.at[index, 'DIED'] = 0
    else:
        df.at[index, 'DIED'] = 1
        
#dropping the old columns, as they serves no more purpose
df.drop(['DATE_DIED', 'SEX', 'PATIENT_TYPE'], axis=1, inplace=True)
```

# First method used: Random Forest Classifier

```python
# making the split for our target(if the patient died)
X = df.drop(['DIED'], axis=1)
y = df['DIED']
```

```python
# making the train and test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.5, random_state=52)
```

```python
# creating our random forest classifier
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=100, random_state=50)
rf.fit(X_train,y_train)
```

```python
#make a prediction and illustrate it's accuracy (or confusion) it in a heatmap
rf_pred=rf.predict(X_test)
cm=confusion_matrix(y_test, rf_pred)
sns.heatmap(cm, annot=True)
plt.show()
```

```python
rf_accuracy=rf.score(X_test, y_test)
rf_accuracy
```

## Try to optimize the score

```python
# sample the dataset as to improve run time
sample_size = 100000

# Randomly sample data points
sampled_df = df.sample(n=sample_size, random_state=42)
```

```python
#making the split again, for the sampled data
s_X = sampled_df.drop(['DIED'], axis=1)
s_y = sampled_df['DIED']

sampled_X_train, sampled_X_test, sampled_y_train, sampled_y_test = train_test_split(s_X, s_y, test_size=.5, random_state=52)
```

```python
#create a numpy array to score results
minE=90
maxE=95
results = np.empty([maxE-minE, 3])
#iterate different number of estimators (ne) and save the scores
for ne in range(minE, maxE):
    rf=RandomForestClassifier(n_estimators=ne, random_state=96)
    rf.fit(sampled_X_train, sampled_y_train)
    trainScore=rf.score(sampled_X_train, sampled_y_train)
    testScore= rf.score(sampled_X_test, sampled_y_test)
    results[ne-minE, 0]=ne
    results[ne-minE, 1]=trainScore
    results[ne-minE, 2]=testScore
```

```python
results_df = pd.DataFrame(results, columns=['Estimators','Train', 'Test'])
```

```python
# check the results
print(results_df['Test'].max(), results_df['Test'].idxmax())

print(results_df.loc[results_df['Test'].idxmax()])
```

#### best result with 93 estimators

```python
#applying the estimators to the entire dataframe
rf2=RandomForestClassifier(n_estimators=93, random_state=50)
rf2.fit(X_train, y_train)
rf2_pred=rf2.predict(X_test)
cm2=confusion_matrix(y_test, rf2_pred)
rf_accuracy= rf2.score(X_test, y_test)
rf_accuracy
```

### although the score improved, it was only slightly
